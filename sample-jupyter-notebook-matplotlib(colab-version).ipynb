```python
# Multiple axes inside one same figure... and setting properties... by using multiple axes directly

fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize = (11.5,6))
fig.suptitle("Figure title here")
fig.set_linewidth(6)
fig.set_edgecolor("chocolate")
fig.supxlabel("X label here")
fig.supylabel("Y label here")
fig.set_facecolor("oldlace")
fig.set_figwidth(11.5)
fig.set_figheight(6)
fig.set_layout_engine(layout="constrained")
ax1.plot(numpy_array, np.sin(numpy_array), color="red", linestyle="dashed")
ax2.scatter(numpy_array[:20], np.exp(numpy_array[:20]), color="orange")
ax3.barh(list(product_dictionary.keys()), list(product_dictionary.values()), color="darkblue")
ax4.hist(sample_array, color="purple")
ax1.set_facecolor("lightgreen")
ax1.set_xmargin(0.1)
ax1.set_ymargin(0.2)
ax1.set_title("Axe 1 title here")
ax2.set_facecolor("lightyellow")
ax2.set_xmargin(0.1)
ax2.set_ymargin(0.2)
ax2.set_title("Axe 2 title here")
ax3.set_facecolor("lightblue")
ax3.set_xmargin(0.1)
ax3.set_ymargin(0.2)
ax3.set_title("Axe 3 title here")
ax4.set_facecolor("pink")
ax4.set_xmargin(0.1)
ax4.set_ymargin(0.2)
ax4.set_title("Axe 4 title here")
plt.show()
```

## Creating plots from Pandas Series and Matplotlib

```python
# preparing Series to plot
car_sales = pd.read_csv("https://raw.githubusercontent.com/danielpm1982/sample-jupyter-notebook-matplotlib/master/car-sales.csv")
# getting Series 'Make' for labels
car_sales_Make_Series = car_sales["Make"]
# getting Series 'Price' for values
car_sales_Price_Series = car_sales["Price"]
# eliminating symbols for plotting
car_sales_Price_Series = car_sales_Price_Series.replace('\$|,', '', regex=True) 
# turning into float type (instead of String) for plotting
car_sales_Price_Series = car_sales_Price_Series.astype("float")
# changing Series index to 'car car_number' for labels
car_sales_Price_Series.index = [f"car {car_number+1}" for car_number in car_sales_Price_Series.index]
# adding the corresponding 'Make' to the 'car car_number' index for labels
car_sales_Price_Series.index = [i + f" ({j})" for i, j in zip(list(car_sales_Price_Series.index), list(car_sales_Make_Series))]
car_sales_Price_Series
```

```python
# setting a default autopct function for pie plot y labels
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n(${v:.2f})'.format(p=pct,v=val)
    return my_autopct

# plotting a pie plot with x (car description) and y (% and price) labels
ax = car_sales_Price_Series.plot(kind="pie", figsize=(13,7), ylabel="", autopct=make_autopct(car_sales_Price_Series.values))
ax.get_figure().suptitle("Pie Chart from Pandas Series with Labels",fontsize=16, fontweight="bold", x=0.52, y=0.92)
ax.get_figure().savefig("./sample-pie-plot.png")
plt.show()
```

## Creating plots from Pandas DataFrames and Matplotlib

```python
# import datetime module and class separately for avoiding conflicts or misuse, as both module and class have the same name, by default
import datetime as date_time_module
from datetime import datetime as date_time_class
import re

# function for extracting string patterns of date (e.g. "1970-01-31") and converting that to a valid datetime date
def extract_date_from_string(date_string_input):
    match = re.search(r'\d{4}-\d{2}-\d{2}', date_string_input)
    date = date_time_class.strptime(match.group(), '%Y-%m-%d').date()
    return date

# function for getting any number of random dates, as a list of datetime dates, considering a customizable date range
# if custom dates are not defined, chooses start date as "1970-01-01", end date as the current date and the no. of samples as only 1 date
# custom input dates (start and end dates) can be passed as a string ("YYYY-MM-DD") or as a datetime date object
def get_random_date_list(start_date="1970-01-01", end_date=date_time_class.now(), number_of_date_samples=1):
    date_list = []
    start_date = extract_date_from_string(str(start_date))
    end_date   = extract_date_from_string(str(end_date))
    num_days   = (end_date - start_date).days
    for i in range(number_of_date_samples):
        rand_days   = np.random.randint(1, num_days)
        random_date = start_date + date_time_module.timedelta(days=rand_days)    
        date_list.append(random_date)
    return date_list

# adding "Sale_Date" (with random dates) and "Total_Sale" with cumsum() to create a cumulative Series
car_sales = pd.read_csv("https://raw.githubusercontent.com/danielpm1982/sample-jupyter-notebook-matplotlib/master/car-sales.csv")
car_sales["Sale_Date"] = get_random_date_list(start_date="2023-01-01", number_of_date_samples=len(car_sales))
car_sales_sorted_dataframe = car_sales.sort_values(by="Sale_Date")
car_sales_sorted_dataframe.reset_index(drop=True, inplace=True)
car_sales_sorted_dataframe["Total_Sale"] = car_sales_sorted_dataframe["Price"].replace("\$|,","",regex=True).astype(float).cumsum()
car_sales_sorted_dataframe
```

```python
car_sales_sorted_dataframe.plot(kind="area", x="Sale_Date", y="Total_Sale", figsize=(8,5), title="Area Chart from Pandas Dataframe with Labels");
```

```python
# Creating a Bar plot from a Pandas DataFrame generated from a NumPy 2D ndarray

# create a NumPy 2D (d0=10, d1=4) ndarray with random float values
random_float_2D_ndarray = np.random.rand(10,4)

# multiply the ndarray float values by 100 and turn them into an int (without using lambda)
# random_int_2D_ndarray = np.multiply(random_float_2D_ndarray, 100).astype(int)

# multiply the ndarray float values by 100 and turn them into an int (using lambda)
multiply_by_100_as_int = lambda x: np.multiply(x, 100).astype(int)
random_int_2D_ndarray = np.array(list(map(multiply_by_100_as_int, random_float_2D_ndarray)))

# create a Pandas Dataframe from the NumPy ndarray with random int values, setting columns and indexes with custom labels
random_int_dataframe = pd.DataFrame(data=random_int_2D_ndarray, columns=["C1", "C2", "C3", "C4"], 
                                    index=["L1", "L2", "L3", "L4", "L5", "L6", "L7", "L8", "L9", "L10"])

random_int_dataframe
```

## Pyplot method (either from plt alias or Pandas) x OO method (creation of fig and ax directly)

```python
# Creating plots with Pandas: pyplot method x OO method

heart_disease = pd.read_csv("https://raw.githubusercontent.com/danielpm1982/sample-jupyter-notebook-matplotlib/master/heart-disease.csv")
heart_disease
```

```python
# pyplot method "plot()" is called at the plt alias or at the DataFrame object, and for simplicity all parameters 
# are passed inside the parenthesis.
# pyplot method is meant for fast previews of what's being plotted, as some of the features are not present and the plot generated 
# is not as complete as it could be. 
# For complete or complex views, use the OO method, by creating fig and axes manually (not implicitly).
# Then, having the fig and axe objects, more customizations and settings are possible than only with the pyplot method.
# See both ways below:

#pyplot method called on a DataFrame object:
heart_disease.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,5), title="Scatter Graph from Pandas using pyplot method");
```

```python
# OO method using subplots() on plt alias and setting fig and ax manually. 
# More complete possibilities of manipulating the plot, for example, by setting the x values range. 
# Observe that the plot() method from the plt alias is never used, but the corresponding plotting method (scatter()), directly 
# on the ax object... returning a PathCollection.
# settings also are set directly on ax and fig objects, with more options available.
fig, ax = plt.subplots(figsize=(8,5))
fig.set_facecolor("lightgreen")
fig.set_layout_engine(layout="constrained")
scatter_path_collection = ax.scatter(x=heart_disease["age"], y=heart_disease["chol"], c=heart_disease["target"])
ax.set_xlim([20,  85])
ax.set_title("Scatter Graph from Pandas using OO method", pad=15)
ax.set_xlabel("Age", labelpad=15)
ax.set_ylabel("Cholesterol", labelpad=15)
ax.set_facecolor("lightyellow")
ax.legend(*scatter_path_collection.legend_elements(), title="Target", loc="upper right")
ax.axhline(heart_disease["chol"].mean(), linestyle = '--')
ax.axhline(heart_disease["chol"].median(), linestyle = ':')
fig.savefig("sample-scatter-plot.png")
plt.show()

# Both approaches could be mixed, for example by calling the below, after creating the fig and ax objects and setting them manually.
# In this case, the ax is set at the plot() method at the DataFrame object as a param arg.
# heart_disease.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,5), title="Scatter Graph... OO method", ax=ax);

```

```python
# Creating plots with Pandas: OO method and multiple axes

heart_disease = pd.read_csv("https://raw.githubusercontent.com/danielpm1982/sample-jupyter-notebook-matplotlib/master/heart-disease.csv")
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(11,8), sharex=True)
fig.suptitle("Multiple-Axe Scatter Graph from Pandas using OO method", fontsize=16, fontweight="bold")
# fig.set_layout_engine("constrained")
scatter_path_collection_0 = ax0.scatter(x=heart_disease["age"], y=heart_disease["chol"], c=heart_disease["target"])
ax0.set(title="Age x Cholesterol x Heart Disease", xlabel="", ylabel="Cholesterol")
ax0.legend(*scatter_path_collection_0.legend_elements(), title="Target")
ax0.axhline(y=heart_disease["chol"].median(), linestyle="--")
scatter_path_collection_1 = ax1.scatter(x=heart_disease["age"], y=heart_disease["thalach"], c=heart_disease["target"])
ax1.set(title="Age x Max Heart Rate x Heart Disease", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*scatter_path_collection_1.legend_elements(), title="Target")
ax1.axhline(y=heart_disease["thalach"].median(), linestyle="--")
ax1.get_figure().savefig("./sample-multiple-axe-plot-3.png")
plt.show()
```
